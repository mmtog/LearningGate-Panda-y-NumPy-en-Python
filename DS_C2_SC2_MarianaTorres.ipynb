```python
#1
# En este primer bloque se  carga el archivo "country_vaccinations.csv",
# convertir la columna de fechas al tipo datetime y mostrar información básica
# del conjunto de datos: número de filas, columnas, rango de fechas y cantidad de países.
import pandas as pd
import numpy as np

filename = 'country_vaccinations.csv'
df = pd.read_csv(filename)
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print("Archivo cargado correctamente")
print("Filas:", len(df), "| Columnas:", len(df.columns))
print("Rango de fechas:", df['date'].min(), "→", df['date'].max())
print("Países distintos:", df['country'].nunique())
df.head()
```

```python
#2
# En este bloque se requiere explorar la información general del DataFrame,
# identificando los tipos de datos, los valores nulos más frecuentes y
# mostrando las primeras filas para una revisión visual inicial.
print("Información general")
print(df.info())

print("Nulos principales")
print(df.isna().sum().sort_values(ascending=False).head(10))

print("Primeras filas")
display(df.head(5))
```

```python
#3
# En este bloque se pide realizar cálculos estadísticos y agrupaciones:
# - Llenar valores nulos en la columna "daily_vaccinations".
# - Obtener el total mundial de vacunaciones.
# - Calcular el promedio global y por país.
# - Determinar el total de dosis aplicadas el 29 de enero de 2021.
# - Calcular los días entre la primera y la última fecha registrada.
# - Mostrar el top 10 de combinaciones de vacunas con mayor número de dosis.
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(0)
vacunas_por_combo = (
    df.groupby('vaccines', as_index=False)['daily_vaccinations']
      .sum()
      .sort_values('daily_vaccinations', ascending=False)
)

total_mundial = df['daily_vaccinations'].sum()

promedio_por_pais = (
    df.groupby('country', as_index=False)['daily_vaccinations']
      .mean()
      .sort_values('daily_vaccinations', ascending=False)
)
promedio_global = promedio_por_pais['daily_vaccinations'].mean()

fecha_obj = pd.Timestamp('2021-01-29')
total_29_enero = df.loc[df['date'] == fecha_obj, 'daily_vaccinations'].sum()

delta_dias = (df['date'].max() - df['date'].min()).days

print(f"Total mundial: {total_mundial:,.0f}")
print(f"Promedio global por país: {promedio_global:,.2f}")
print(f"Total aplicado el 29/01/2021: {total_29_enero:,.0f}")
print(f"Días entre fechas: {delta_dias}")
print("Top 10 combinaciones de vacunas")
display(vacunas_por_combo.head(10))
```

```python
###4
# En este último bloque final se solicita crear nuevas columnas para análisis adicionales:
# - "diferencias": resta entre vacunaciones diarias ajustadas y sin ajustar.
# - "canVac": número de tipos de vacunas distintas utilizadas por país.
# Además, se filtran subconjuntos del DataFrame:
# - Registros antes del 20 de diciembre de 2020.
# - Registros que contienen la vacuna Pfizer.
# Finalmente, se imprimen los tamaños de los nuevos DataFrames
# y se muestran las primeras filas de cada uno.
df['daily_vaccinations_raw'] = df['daily_vaccinations_raw'].fillna(0)
conDiferencias = df.copy()
conDiferencias['diferencias'] = (
    conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
)

conCantidad = df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].apply(
    lambda v: 0 if pd.isna(v) else len([m.strip() for m in str(v).split(',') if m.strip()])
)

antes20 = df[df['date'] < pd.Timestamp('2020-12-20')].copy()
pfizer = df[df['vaccines'].str.contains('Pfizer', case=False, na=False)].copy()

print("Filas conDiferencias:", conDiferencias.shape[0])
print("Filas conCantidad:", conCantidad.shape[0])
print("Filas antes20:", antes20.shape[0])
print("Filas pfizer:", pfizer.shape[0])

display(conDiferencias.head(3))
display(conCantidad.head(3))
display(antes20.head(3))
display(pfizer.head(3))
```
